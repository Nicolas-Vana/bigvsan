```python
import subprocess
from pathlib import Path

def convert_opus_to_wav(source_folder, output_folder):
    """
    Converts all .opus files in the given folder to .wav format, maintaining the folder structure.
    Skips conversion if the .wav file already exists in the target folder.

    :param source_folder: Folder containing .opus files.
    :param output_folder: Root folder to save the converted .wav files, maintaining the source structure.
    """
    source_path = Path(source_folder)
    output_path = Path(output_folder)

    for opus_file in source_path.rglob('*.opus'):
        relative_path = opus_file.relative_to(source_path)
        target_subfolder = output_path / relative_path.parent
        target_subfolder.mkdir(parents=True, exist_ok=True)

        wav_file = target_subfolder / relative_path.with_suffix('.wav').name
        print(wav_file)

        if not wav_file.exists():
            subprocess.run(['ffmpeg', '-i', str(opus_file), str(wav_file)], stdout=subprocess.PIPE, stderr=subprocess.PIPE)
        else:
            print(f"Skipping conversion, file already exists: {wav_file}")

# Example usage
source_folder = './mls_portuguese_opus/train/audio'  # Unix-style path
output_folder = './mls_portuguese_wav/train/audio'   # Unix-style path for output
convert_opus_to_wav(source_folder, output_folder)
```
